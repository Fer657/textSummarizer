```python
def get_product(x:int,y:int) -> int:
    return x*y
```

```python
get_product(x =2,y=5)
```

```python
get_product(x =2,y="5") # Still runs even if string parameters therefore we use ensure anotation
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int,y:int) -> int:
    return x*y

get_product(x =2,y="5") # wont work now
 
```
